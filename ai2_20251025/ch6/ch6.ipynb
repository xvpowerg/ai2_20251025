## Sympy常用方法

1. 基本符號操作

x, y = sp.symbols('x y')

2. 代數運算

expand(): 展開多項式表達式。

factor(): 因式分解多項式表達式。

simplify(): 簡化數學表達式。

3. 代數方程求解

solve(): 解代數方程或方程組。

diff(): 求導數，可以求高階導數

integrate(): 積分，可以計算定積分和不定積分。





### 練習1：代數展開
公式：
$(x + y)^3$

```python
import sympy as sp
from IPython.display import display, Math
x,y = sp.symbols("x y")
expr1 = (x + y ) ** 3
expand_expr1 = sp.expand(expr1)
print(expand_expr1,sp.latex(expand_expr1))

display(Math(f"({x}+{y})^3"))
display(Math(sp.latex(expand_expr1)))
```

### 練習2：因式分解
公式：
$[ x^3 + 3x^2 + 3x + 1 ]$

```python
x = sp.symbols("x")
expr2 = x **3 + 3 *ｘ**2 +3 * x + 1
f_expr2 = sp.factor(expr2)
display(Math(rf"{sp.latex(expr2)}"))
display(Math(rf"{sp.latex(f_expr2)}"))
```

### 練習3：代數簡化
公式：
$[ \frac{x^2 - y^2}{x - y} ]$

```python
x,y = sp.symbols("x y")
expr = (x **2 - y **2) / (x - y)
result = sp.simplify(expr)
display(Math(rf"{sp.latex(expr)}"))
display(Math(rf"{sp.latex(result)}"))
```

### 練習4：符號代入
公式：
$[ x^2 + y^2 ]$
並代入 $( x = 1 )$ 和 $( y = 2 )$

```python
import sympy.core.add as add
x,y = sp.symbols("x y")
expr4:add.Add = x ** 2 + y ** 2

print(type(expr4))
sub_expr4 = expr4.subs(x,1).subs(y,2)
display(Math(fr"{sp.latex(expr4)}"))
display(Math(fr"{sp.latex(sub_expr4)}"))
```

### 練習5：求根
公式：
$[ x^2 - 4 = 0 ]$

```python
x = sp.symbols("x")
expr5 = x ** 2 - 4
root_expr5 = sp.solve(expr5)
display(Math(fr"{sp.latex(expr5)}"))
display(Math(fr"{sp.latex(root_expr5)}"))
```


### 練習1：二元線性方程組
#### 公式
1. \( x + 2y = 3 \)
2. \( 3x - y = 4 \)

```python
x,y = sp.symbols("x y")
eq1_1 = sp.Eq(x + 2 * y,3)
eq1_2 = sp.Eq(3 *x - y ,4)
result = sp.solve((eq1_1,eq1_2),(x,y))
display(eq1_1)
display(eq1_2)
print(result)
print("x=",result[x],"y=",result[y])
```

## 微分

- $ f(x) = \text{常數} \implies f'(x) = 0 $

- 次方：$ f(x) = x^n \implies f'(x) = nx^{n-1} $

- 函數：$ f(x) = Cg(x) \implies f'(x) = Cg'(x) $

- 兩個函數：$ f(x) = g(x) + h(x) \implies f'(x) = g'(x) + h'(x) $

- 乘積：
  $ \frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x) $

- 商：
  $ r(x) = \frac{s(x)}{t(x)} \implies r'(x) = \frac{s'(x)t(x) - s(x)t'(x)}{[t(x)]^2} $

- 連鎖律(Chain Rule)：
  $ \frac{d}{dx}[o(i(x))] = o'(i(x)) \cdot i'(x) $



### 練習1：一般微分
$ f(x) = x^4 + 3x^3 + x^2 + x + 1  $

```python
import sympy as sp
from IPython.display import display, Math
x = sp.symbols("x")
f1 = x**4 + 3 * x ** 3 + x ** 2 + x + 1
f1_diff = sp.diff(f1,x)
print(sp.latex(f1_diff))
display(Math(fr"f(x)={sp.latex(f1)}"))
display(Math(fr"f(x)={sp.latex(f1_diff)}"))
```

## 偏微分

$ f(x, y) = x^2 + 3xy + y^2 $

## 不定積分

$ \int (x^2 + 3x + 2) \, dx $

```python
import sympy as sp
from IPython.display import display, Math

# 定義變量
x = sp.symbols('x')
# 不定積分
f1 = x**2 + 3*x + 2
integral_f1 = sp.integrate(f1,x)
display(Math(sp.latex(integral_f1) ))
display(Math(fr'\int ({sp.latex(f1)}) \, dx = {sp.latex(integral_f1) + " + C"}'))
```

### 定積分
$ \int_0^1 (x^2 + 3x + 2) \, dx $

```python
# 定義變量
x = sp.symbols('x')
# 定積分
f2 = x**2 + 3*x + 2
integral_f2 = sp.integrate(f2,(x,0,1))
display(Math(fr'\int_0^1 ({sp.latex(f2)}) \, dx = {sp.latex(integral_f2)}'))
```

# 🏡 Ames Housing 欄位名稱對照表

### 識別資訊

* **PID**：地號（Parcel Identification Number）
* **MS SubClass**：建物等級（建物類型編碼，例如 20=1層單戶住宅，60=2層新型單戶住宅…）
* **MS Zoning**：建築分區（如 RL=住宅低密度，RM=住宅中密度）

### 地段與土地

* **Lot Frontage**：臨街長度（英尺）
* **Lot Area**：地坪面積（平方英尺）
* **Street**：街道型態（Pave=柏油，Grvl=碎石）
* **Alley**：巷道型態（Grvl=碎石，Pave=柏油，Na=無）
* **Lot Shape**：地形形狀（Reg=規則，IR1/IR2/IR3=不規則）
* **Land Contour**：地勢（Lvl=平坦，Bnk=坡地，HLS=小山丘，Low=低窪）
* **Utilities**：公用設施（AllPub=水電瓦斯全有）
* **Lot Config**：地塊配置（Inside=一般，Corner=角地，CulDSac=死巷，FR2/FR3=臨街）
* **Land Slope**：坡度（Gtl=緩坡，Mod=中等，Sev=陡坡）
* **Neighborhood**：社區（如 NAmes, Gilbert, StoneBr 等）
* **Condition 1**：近主幹道/鐵路/其他條件（Norm=正常，Feedr=臨支路，PosA/PosN=臨公園/綠地…）
* **Condition 2**：第二近的條件（同上，通常大多數是 Norm）

### 建物類型與風格

* **Bldg Type**：建築類型（1Fam=單戶，TwnhsE=聯排端戶，Duplex=雙戶…）
* **House Style**：屋型（1Story=平房，2Story=兩層，SLvl=分層，SFoyer=挑高玄關）

### 整體評分

* **Overall Qual**：整體品質（1–10，越高越好）
* **Overall Cond**：整體狀況（1–10）

### 建造資訊

* **Year Built**：建造年份
* **Year Remod/Add**：翻修/增建年份

### 屋頂與外牆

* **Roof Style**：屋頂型態（Gable=山形，Hip=四坡屋頂…）
* **Roof Matl**：屋頂材質（CompShg=瓦片，WdShng/WdShake=木瓦…）
* **Exterior 1st**：外牆材質1
* **Exterior 2nd**：外牆材質2
* **Mas Vnr Type**：外牆飾面材質（BrkFace=磚，Stone=石頭，None）
* **Mas Vnr Area**：外牆飾面面積（平方英尺）

### 外觀品質

* **Exter Qual**：外部品質（Ex/Gd/TA/Fa/Po）
* **Exter Cond**：外部狀況（Ex/Gd/TA/Fa/Po）

### 基礎與地下室

* **Foundation**：地基類型（PConc=混凝土，CBlock=水泥磚…）
* **Bsmt Qual**：地下室高度品質
* **Bsmt Cond**：地下室狀況
* **Bsmt Exposure**：地下室採光（Gd=好，Mn=一般，No=無，Av=平均）
* **BsmtFin Type 1**：地下室完工類型1（GLQ=良好居住，ALQ=平均居住，Unf=未完成…）
* **BsmtFin SF 1**：地下室完工面積1（平方英尺）
* **BsmtFin Type 2**：地下室完工類型2
* **BsmtFin SF 2**：地下室完工面積2
* **Bsmt Unf SF**：地下室未完工面積
* **Total Bsmt SF**：地下室總面積

### 暖氣與冷氣

* **Heating**：供暖系統（GasA=天然氣，Wall=壁掛…）
* **Heating QC**：供暖品質（Ex/Gd/TA/Fa/Po）
* **Central Air**：中央空調（Y/N）
* **Electrical**：電氣系統（SBrkr=斷路器，FuseA=保險絲…）

### 地上面積

* **1st Flr SF**：一樓面積
* **2nd Flr SF**：二樓面積
* **Low Qual Fin SF**：低品質完工面積

### 浴室與臥室

### 為什麼需要 OneHotEncoder

1. **轉換類別變數**：
   - 類別變數不能直接用於大多數機器學習模型，需要將其轉換為數值型。
   - OneHotEncoder 將每個類別變數轉換為多個二進位變數，使模型能夠理解並利用這些特徵。

2. **避免引入順序關係**：
   - 將類別變數轉換為數字（如使用 LabelEncoder）可能會引入不必要的順序關係，而 OneHotEncoder 避免了這一點。
   - 保證了每個類別之間的獨立性，不會因為數值大小而影響模型的預測。

3. **提高模型性能**：
   - 對於很多模型來說，使用 OneHotEncoder 編碼後的特徵可以提高模型的性能，因為它能更好地表示數據的結構和模式。

在數據分析和機器學習中，數據類型可以大致分為以下幾種，每種都有其特定的用途和處理方式：

1. **類別數據（Categorical Data）**
   - **說明**：表示不同類別或群組的數據，通常是非數值的。
   - **案例**：性別（男、女）、血型（A、B、AB、O）、產品類型（家具、電子產品、服裝）。

2. **數值數據（Numerical Data）**
   - 分為兩種：
     - **連續數據（Continuous Data）**
       - **說明**：可以在一定範圍內任意細分的數據。
       - **案例**：身高、體重、溫度、股票價格。
     - **離散數據（Discrete Data）**
       - **說明**：具體的整數值，不可以細分。
       - **案例**：班上學生人數、公司員工數、一天中的客戶服務電話次數。

3. **時間數據（Time Series Data）**
   - **說明**：按時間順序排列的數據點序列，通常每個數據點都有一個時間戳。
   - **案例**：每日股市收盤價、每小時天氣記錄、月銷售數據。

4. **文字數據（Text Data）**
   - **說明**：由字、詞、句子構成的數據，通常需要特殊的處理方式如自然語言處理（NLP）。
   - **案例**：商品評論、新聞文章、社交媒體帖子。

5. **布爾數據（Boolean Data）**
   - **說明**：表示是或否的二元數據。
   - **案例**：是否已婚（是、否）、是否訂閱電子報（是、否）。

6. **序列數據（Ordinal Data）**
   - **說明**：類別數據的一種，但類別之間存在自然排序或等級。
   - **案例**：滿意度調查（不滿意、一般、滿意、非常滿意）、教育程度（小學、中學、大學、研究生）。

- **85**	SPLIT FOYER 分割門廳
- **90**	DUPLEX - ALL STYLES AND AGES 雙拼 - 所有風格和年齡
- **120**	1-STORY PUD (Planned Unit Development) - 1946 & NEWER 一層樓PUD（計劃單元開發）- 1946年及以後
